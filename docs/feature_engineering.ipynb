## basic features - relative positions

We first construct graphs using redshift space positions by taking a $(2 \textrm{ Mpc}, 10 \textrm{ Mpc})$ cylinder around central galaxies and connection all "satellite" nodes to the center. The use of the word satellite here does not refer to the real central/satellite distinction, but we want to remain impervious to those definitions. We can remain confident that our choices of centrals within galaxy surverys should thus not matter as we will always just consider every galaxy about one galaxy of reference.

Since we want the model to be equivariant, we began with an incredibly simple fiducial graph:

Nodes:
$$N=\{r_\perp - \bar{r}_{\perp}, r_{\rm LOS}-\bar{r}_{\rm LOS}\}$$
Edges:
$$E=\{s, \mu\}$$

Which are the two standard ways to decompose redshift space positions by observations except we center the node positions with respect to the graph center. The center node is connected to all nodes, while satellites are only connected to neighbors within a region.

The graphs are roughly based off of the CiC (**CITE!!!**) formalism
